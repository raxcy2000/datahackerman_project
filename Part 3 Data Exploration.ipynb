```python
lat_long = []
for i, j in zip(data["location.latitude"][0:1000], data["location.longitude"][0:1000]):
    lat_long.append((i, j))
```

```python
# Make an empty map
m = folium.Map(location=lat_long[0], tiles="OpenStreetMap", zoom_start=2)

# add marker one by one on the map
for i in range(0,len(data[0:1000])):
    folium.Marker(
      location=[data.iloc[i]["location.latitude"], data.iloc[i]["location.longitude"]],
      popup=data.iloc[i]["location.country"],
   ).add_to(m)

# Show the map again
m
```
